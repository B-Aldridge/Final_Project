## Could not find original notebook for data cleaning. This is an example of the workflow

```python
# Load the data
quotes = pd.read_csv("quotes.csv")
actors = pd.read_csv("actors_cleaned.csv")
```

```python
# Change column names
actors.rename(columns={'MOVIE_TITLE': 'movie_title'}, inplace=True)
quotes.rename(columns={'TEXT': 'quote'}, inplace=True)
quotes.rename(columns={'CHARACTER': 'character'}, inplace=True)
actors.rename(columns={'primaryName': 'actor'}, inplace=True)
actors.rename(columns={'characters': 'character'}, inplace=True)
quotes.rename(columns={'MOVIE_TITLE': 'movie_title'}, inplace=True)
actors.rename(columns={'originalTitle': 'movie_title'}, inplace=True)

```

```python
quotes.head()
```

```python
actors.head()
```

```python
# Drop the unused columns
actors = actors.drop(columns=["Unnamed: 0"])
actors = actors.drop(columns=["primaryTitle"])
```

```python
actors.head()
```

```python
quotes.head()
```

```python
# Convert names to lowercase for comparison
quotes["character"] = quotes["character"].str.strip().str.lower()
actors["character"] = actors["character"].str.strip().str.lower()
actors["movie_title"] = actors["movie_title"].str.strip().str.lower()
quotes["quote"] = quotes["quote"].str.strip().str.lower()
```

```python
# Merge the dataframes
merged_data = quotes.merge(actors, on="character", how="left")

# Display the merged dataframe
merged_data.head(5)
```

```python
merged_data
```

```python
# Remove rows with NaN values
merged_data_cleaned = merged_data.dropna()

# Display the cleaned dataframe
merged_data_cleaned
```

```python
# Drop duplicate rows
quotes_cleaned = merged_data_cleaned.drop_duplicates(subset=["quote"], keep="first")

# Display the cleaned dataframe
quotes_cleaned.head()
```

```python
quotes_cleaned
```

```python
# Reset the index and remove "movie_title_y" column
quotes_cleaned = quotes_cleaned.drop(columns=["movie_title_y"]).reset_index(drop=True)

# Display the cleaned dataframe
print(quotes_cleaned.head())
```
