```python
#import dependencies
import pandas as pd
```

```python
#import data
title_principals_df=pd.read_csv('title_principals.tsv', sep='\t')
```

```python
names_basics_df=pd.read_csv('name_basics.tsv', sep='\t')
```

```python
title_basics_df=pd.read_csv("title_basics.tsv", sep='\t')
```

```python
#merge data
df=title_principals_df.merge(names_basics_df)
```

```python
df=df.merge(title_basics_df)
```

```python

df.head(5)
```

```python
#select desired columns with information for making the model and for filtering
df=df[["category","characters","primaryName","titleType","primaryTitle","originalTitle","startYear","genres"]]
```

```python
df.head(5)
```

```python
df['category'].value_counts()
```

```python
#filter only actor roles
cleaned_df=df[df['category'].isin(['actor','actress'])]
```

```python
cleaned_df['category'].value_counts()
```

```python
cleaned_df
```

```python
cleaned_df["titleType"].unique()
```

```python
#filter only movies
cleaned_df=cleaned_df[cleaned_df['titleType'].isin(['movie'])]
```

```python
#filter only roles after 1912
cleaned_df=cleaned_df.loc[cleaned_df['startYear']!='\\N']
cleaned_df["startYear"]=cleaned_df["startYear"].astype(int)
```

```python
cleaned_df=cleaned_df.loc[cleaned_df['startYear']>1912]
```

```python
#Delete rows with "\N" for characters column
cleaned_df=cleaned_df.loc[cleaned_df['characters']!='\\N']
```

```python
#reset the index since many rows were removed
cleaned_df=cleaned_df.reset_index(drop='True')
```

```python
#clean the character column to only show the character name
cleaned_df['characters']=cleaned_df['characters'].str.replace("\[\"","")
cleaned_df['characters']=cleaned_df['characters'].str.replace("\"\]","")
```
